Implementasi Singly Linked List

```python
# Definisi node dalam Linked List
class TaskNode:
  def __init__(self, task_name):
    self.task_name = task_name # Menyimpan nama tugas
    self.next = None # Pointer ke tugas berikutnya

# Definisi Linked List untuk Daftar Tugas
class TaskList:
  def __init__(self):
    self.head = None #Awalnya tidak ada tugas (daftar kosong)

  # Menambahkan tugas ke Linked List
  def add_task(self, task_name):
    new_task = TaskNode(task_name) # Membuat node baru untuk tugas
    if not self.head:
      self.head = new_task # Jika kosong, tugas pertama menjadi head
    else:
      current = self.head
      while current.next: # Mencari node terakhir dalam list
        current = current.next
      current.next = new_task #menambahkan tugas di akhir list

  # Menampilkan daftar tugas
  def show_tasks(self):
    if not self.head:
      print("\n Tidak ada tugas hari ini!")
      return

    print("\n Daftar Tugas Hari Ini:")
    current = self.head
    count = 1
    while current:
      print(f"{count}. {current.task_name}")
      current = current.next
      count += 1

# ====== Program Utama (Main) ====== #

# Membuat objek daftar tugas
task_list = TaskList()

# Meminta pengguna memasukkan jumlah tugas
num_tasks = int(input("Masukkan jumlah tugas hari ini: "))

# Loop untuk memasukkan tugas ke dalam Linked List
for i in range(1, num_tasks + 1):
  task_name = input(f"Masukkan tugas ke-{i}: ")
  task_list.add_task(task_name)

# Menampilkan daftar tugas
task_list.show_tasks()
```

Implementasi Doubly Linked List

```python
# Definisi Node dalam Doubly Linked List
class PageNode:
  def __init__(self, url):
    self.url = url # URL alamat web
    self.next = None # Pointer ke halaman berikutnya
    self.prev = None # Pointer ke halaman sebelumnya

# Definisi Doubly Linked List untuk Riwayat Peramban
class BrowserHistory:
  def __init__(self):
    self.current = None # Pointer ke halaman yang sedang dibuka

  # Menambahkan halaman baru ke dalam riwayat
  def visit_page(self, url):
    new_page = PageNode(url)
    if not self.current:
      self.current = new_page #Jika riwayat kosong, halaman ini akan menjadi halaman pertama.
    else:
      new_page.prev = self.current
      self.current.next = new_page
      self.current = new_page # Pindahkan posisi ke halaman baru

  # Memanmpilkan riwayat kunjungan
  def show_history(self):
    if not self.current:
      print("\n Riwayat kosong.")
      return

    print("\n Riwayat peramban Web:")
    temp = self.current
    while temp.prev: # Kembali ke halaman pertama
      temp = temp.prev

    count = 1
    while temp:
      print(f"{count}. {temp.url}")
      temp = temp.next
      count += 1

  # Navigasi mundur ke halaman sebelumnya
  def go_back(self):
    if self.current and self.current.prev:
      self.current = self.current.prev
      print(f"\n Anda kembali ke: {self.current.url}")
    else:
      print(f"Tidak bisa kembali, ini adalah halaman pertama!")

  # Navigasi maju menuju halaman berikutnya
  def go_forward(self):
    if self.current and self.current.next:
      self.current = self.current.next
      print(f"\n Anda maju ke: {self.current.url}")
    else:
      print("\n Tidak bisa maju, ini adalah halaman terakhir!")

# ===== Program Utama (Main) ===== #

# Membuat objek riwayat peramban
history = BrowserHistory()

# Meminta pengguna memasukkan jumlah halaman yang dikunjungi
num_pages = int(input("\n Masukkan jumlah halaman yang dikunjungi"))

# Looop untuk memasukkan URL halaman ke dalam Doubly Linked List
for i in range(1, num_pages + 1):
  url = input(f"Masukkan URL halaman ke-{i}: ")
  history.visit_page(url)

# Menampilkan riwayat yang telah dimasukkan
history.show_history()

# Menu navigasi pengguna
while True:
  command = input("\n Tekan 'b' untuk kembali, 'f' untuk maju, atau 'x' untuk keluar. \nMasukkan perintah: ").lower()

  if command == 'b':
    history.go_back()
  elif command == 'f':
    history.go_forward()
  elif command == 'x':
    print("\n Keluar dari riwayat peramban.")
    break
  else:
    print("\n Perintah tidak dikenali.")
```

Implementasi Circular Linked List

```python
# Definisi Node dalam Circular Linked List
class SongNode:
  def __init__(self, title):
      self.title = title # Menyimpan judul lagu
      self.next = None # Pointer ke Lagu berikutnya

# Definisi Circular Linked List untuk playlist lagu
class Playlist:
  def __init__(self):
    self.current = None # Lagu sedang diputar

  # Menambahkan lagu ke dalam playlist
  def add_song(self, title):
    new_song = SongNode(title)
    if not self.current:
      self.current = new_song
      self.current.next = self.current # Menghubungkan ke dirinya sendiri (Circular)
    else:
      temp = self.current
      while temp.next != self.current: # Mencari lagu terakhir
        temp = temp.next
      temp.next = new_song # Menambahkan lagu terakhir di akhir
      new_song.next = self.current # Menghubungkan kembali ke awal

# Menampilkan daftar lagu dalam playlist
  def show_playlist(self):
    if not self.current:
      print("\n Playlist kosong!")
      return

    print("\n Playlist lagu anda:")
    temp = self.current
    count = 1
    while True:
      print(f"{count}. {temp.title}")
      temp = temp.next
      count += 1
      if temp == self.current: # Berhenti jika sudah kembali ke awal
        break

# Memutar lagu berikutnya dalam playlist
  def next_song(self):
    if self.current:
      self.current = self.current.next
      print(f"\n Memutar: {self.current.title}")

# ====== Program Utama (Main) ====== #

# Membuat objek playlist
playlist = Playlist()

# Meminta pengguna untuk memasukkan jumlah lagu
num_songs = int(input("\n Masukkan jumlah lagu dalam playlist: "))

# Loop untuk memasukkan lagu ke dalam Circular Linked List
for i in range(1, num_songs + 1):
  title = input(f"Masukkan judul lagu ke-{i}: ")
  playlist.add_song(title)

# Menampilkan daftar lagu
playlist.show_playlist()

# Menu navifasi pengguna
while True:
  command = input("\n Tekan 'n' untuk lagu berikutnya, Tekan 'x' untuk keluar. \nMasukkan perintah: ").lower()
  if command == 'n':
    playlist.next_song()
  elif command == 'x':
    print("\n Keluar dari playlist lagu.")
    break
  else:
    print("\n Perintah tidak dikenali.")
```

```python
# Definisi Node dalam Singly Linked List
class Node:
  def __init__(self, data):
      self.data = data # Menyimpan nilai node
      self.next = None # Pointer ke node berikutnya

# Definisi SIngly Linked List dengan Berbagai Operasi
class LinkedList:
  def __init__(self):
      self.head = None # Awalnya, daftar kosong

  # Menambahkan elemen di awal
  def add_at_beginning(self, data):
      new_node = Node(data)
      new_node.next = self.head
      self.head = new_node

  # Menambahkan elemen di akhir
  def add_at_end(self, data):
      new_node = Node(data)
      if not self.head:
        self.head = new_node
        return
      temp = self.head
      while temp.next:
          temp = temp.next
      temp.next = new_node

  # Menambahkan elemen di posisi tertentu
  def add_at_position(self, position, data):
      if position < 0:
        print("Posisi tidak valid!")
        return
      new_node = Node(data)
      if position == 0:
        self.add_at_begining(data)
        return
      temp = self.head
      for _ in range(position - 1):
        if not temp:
          print("Posisi melebihi oanjang daftar!")
          return
        temp = temp.next
      new_node.next = temp.next
      temp.next = new_node

  # Menghapus elemen di awal
  def delete_at_beginning(self):
      if not self.head:
        print("Daftar kosong, taidak ada yang bisa dihapus!")
        return
      self.head = self.head.next

  # Mengahapus elemen di akhir
  def delete_at_end(self):
      if not self.head:
        print("Daftar kosong, tidak ada yang bisa dihapus!")
        return
      temp = self.head
      while temp.next.next:
        temp = temp.next
      temp.next = None

  # Menghapus elemen di posisi tertentu
  def delete_at_position(self, position):
    if not self.head:
      print("Daftar kosong, tidak ada yang bisa dihapus!")
      return
    if position == 0:
      self.head = self.head.next
      return
    temp = self.head
    for _ in range(position - 1):
      if not temp or temp.next:
        print("‚ö†Ô∏èPosisi melebihi panjang daftar!")
        return
      temp = temp.next
    temp.next = temp.next.next

  # Mencari elemen dalam Linked List
  def search(self, key):
    temp = self.head
    while temp:
        if temp.data == key:
          print(f"Elemen {key} ditemukan dalam daftar!")
          return
        temp = temp.next
    print(f"Elemen {key} tidak ditemukan dalam daftar.")

  # Membalik (reverse) Linked List
  def reverse(self):
    prev = None
    current = self.head
    while current:
      next_node = current.next
      current.next = prev
      prev = current
      current = next_node
    self.head = prev

  # Menampilkan Linked List
  def display(self):
    if not self.head:
      print("Daftar kosong!")
      return
    temp = self.head
    while temp:
      print(temp.data, end=" -> ")
      temp = temp.next
    print("None")

# ====== Program Utama (Main) ====== #

# Membuat objek Linked List
ll = LinkedList()

# Meminta pengguna memasukkan elemen awal
num_elements = int(input("\nMasukkan jumlah awal elemen dalam daftar: "))
for i in range(num_elements):
  data = int(input(f"Masukkan data ke-{i+1}: "))
  ll.add_at_end(data)

# Menu Operasi Linked List
while True:
  print("\n Pilih Operasi:")
  print("1. Tambahkan elemen di awal")
  print("2. Tambahkan elemen di akhir")
  print("3. Tambahkan elemen di posisi tertentu")
  print("4. Hapus elemen di awal")
  print("5. Hapus elemen di akhir")
  print("6. Hapus elemen di posisi tertentu")
  print("7. Cari elemen dalam daftar")
  print("8. Balikan daftar")
  print("9. Tampilkan daftar")
  print("0. Keluar")

  choice = int(input("Masukkan pilihan: "))

  if choice == 1:
    data = int(input("Masukkan data yang ingin ditambahkan di awal: "))
    ll.add_at_beginning(data)
  elif choice == 2:
    data = int(input("Masukkan data yang ingin ditambahkan di akhir: "))
    ll.add_at_end(data)
  elif choice == 3:
    position = int(input("Masukkan posisi untuk menambahkan data: "))
    data = int(input("Masukkan data yang ingin ditambahkan: "))
    ll.add_at_position(position, data)
  elif choice == 4:
    ll.delete_at_beginning()
  elif choice == 5:
    ll.delete_at_end()
  elif choice == 6:
    position = int(input("Masukkan posisi elemen yang ingin dihapus: "))
    ll.delete_at_position(position)
  elif choice == 7:
    key = int(input("Masukkan elemen yang ingin dicari: "))
    ll.search(key)
  elif choice == 8:
    ll.reverse()
    print("Daftar berhasil dibalik!")
  elif choice == 9:
    ll.display()
  elif choice == 0:
    print("\nKeluar dari program.")
    break
  else:
    print("Pilihan tidak valid. Silakan coba lagi.")
```

```python
class Node:
    def __init__(self, name):
        self.name = name
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, name):
        new_node = Node(name)
        if not self.head:
            self.head = new_node
            self.head.next = self.head
            self.tail = self.head
        else:
            self.tail.next = new_node
            new_node.next = self.head
            self.tail = new_node

    def print_schedule(self, days):
        if not self.head:
            print("Tidak ada satpam yang bertugas.")
            return

        print("\n‚≠ê Jadwal Piket Satpam:\n")
        temp = self.head
        for i in range(1, days + 1):
            print(f"Hari {i}: {temp.name}")
            temp = temp.next

circular_list = CircularLinkedList()
num_satpam = int(input("Masukkan jumlah satpam yang akan bertugas: "))

satpam_list = []
for i in range(1, num_satpam + 1):
    name = input(f"Masukkan nama satpam ke-{i}: ")
    satpam_list.append(name)
    circular_list.append(name)

print("\n‚úÖ Daftar Satpam yang Bertugas:")
for id, name in enumerate(satpam_list, start=1):
    print(f"{id}. {name}")

num_days = int(input("\nMasukkan jumlah hari piket: "))

circular_list.print_schedule(num_days)
```
